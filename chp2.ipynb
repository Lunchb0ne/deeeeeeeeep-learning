```python
np.random.seed(42)
s1_data = ImageDataBunch.from_folder(path, train='.', valid_pct = 0.2, ds_tfms=get_transforms(), size=224)
s2_data = s1_data.normalize(imagenet_stats)
```

```python
s2_data.classes
```

```python
s2_data.show_batch(rows=3)
```

```python
model = cnn_learner(s1_data, models.resnet34, metrics=error_rate)
```

```python
model.fit_one_cycle(4)
```

```python
model.save('stage-1')
```

```python
model.unfreeze()
```

```python
model.lr_find()
```

```python
model.recorder.plot()
```

```python
model.fit_one_cycle(2,max_lr=slice(8e-5, 9e-4))
```

```python
model.save('stage-2')
```

```python
model.load('stage-2')
```

```python
interp = ClassificationInterpretation.from_learner(model)
```

```python
interp.plot_confusion_matrix(figsize=(5,5))
```

```python
interp.plot_top_losses(9)
```
